```python
def checkTicker(ticker):
    if not isinstance(ticker, str):
        return False
    for c in ticker:
        if c not in 'QWERTYUIOPASDFGHJKLZXCVBNM.':
            return False
    return True
```

```python
def castDate(date):
    try:
        date = pd.Timestamp(date)
    except:
        date = np.datetime64('NaT')
    return date
```
